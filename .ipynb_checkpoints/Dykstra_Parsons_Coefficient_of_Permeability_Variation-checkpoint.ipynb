```python
from matplotlib import pyplot as plt
import seaborn
import probscale
from pylab import *
import numpy as np
x = np.array([0, 5.8, 11.8, 17.7, 23.5, 29.4, 35.3, 41.2, 47.1, 52.9, 58.8, 64.7, 70.6, 76.5, 82.4, 88.2, 94.1])
y = np.array([1212, 900, 730, 714, 650, 591, 565, 500, 450, 440, 430, 420, 407, 381, 324, 315, 283])
fig, ax = plt.subplots(figsize=(10, 8))
ax.set_xlim(0.5, 99)
ax.set_xscale('prob')
ax.set_ylim(1e1, 1e4)
ax.set_yscale('log')
seaborn.despine(fig=fig)
plt.plot(x, y, 'go')
plt.title('Permeability Variation')
plt.ylabel('Permebility, md')
plt.xlabel('Percent of Samples with Larger Permebility')
plt.grid(True)
# Mean for x and y
mean_x = np.mean(x)
mean_y = np.mean(y)

# Total number of values
m = len(x)

# Calculate b1 and b2
numer = 0
denom = 0
for i in range(m):
    numer += (x[i] - mean_x) * (y[i] - mean_y)
    denom += (x[i] - mean_x) ** 2
b1 = numer / denom
b0 = mean_y - (b1 * mean_x)

# Print coefficients
print('b1 = ', b1, 'b0 = ', b0)
# Plot regression line
y_pred = b0 + b1 * x
plt.plot(x, y_pred, color='#58b970', label='Regression Line')
#Calculate permeability at 84.1% probability and at 50% 
#Calculate variance
k1 = b0 + b1*50;
k2 = b0 + b1*84.1
v = (k1 - k2) / k1
print('v = ', v)
#Calculate r^2 score
sumofsquares = 0
sumofresiduals = 0
for i in range(m) :
    y_pred = b0 + b1 * x[i]
    sumofsquares += (y[i] - mean_y) ** 2
    sumofresiduals += (y[i] - y_pred) **2
    
score  = 1 - (sumofresiduals/sumofsquares)
print('R^2 score = ', score)
#Plot v on graph
plt.text(40, 2000, 'v= ', fontsize = 16)
plt.text(47, 2000, round(v, 2), fontsize = 16)
```
